```python
import pandas as pd
# import selenium 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
path = 'D:\Projects\chromedriver'
service = Service(executable_path=path)

# Configuración del driver para Brave
option = webdriver.ChromeOptions()
option.binary_location = 'C:/Program Files/BraveSoftware/Brave-Browser/Application/brave.exe'
driver = webdriver.Chrome(service=service, options=option)

url = "https://www.youtube.com/@BAXBESTIATV/videos"
driver.get(url)
```

style-scope ytd-rich-grid-media -> class of the container  
//\*[@id="dismissible"] -> xpath of the container  

//\*[@id="video-title"] -> xpath video title  
//\*[@id="metadata-line"]/span[1] -> xpath video views  
//\*[@id="video-title-link"] -> xpath video link  

```python
videos = driver.find_elements(By.CLASS_NAME, value='style-scope ytd-rich-grid-media')

videos_list = []
for idx, video in enumerate(videos):
    title = video.find_element(by='xpath', value='.//*[@id="video-title"]').text
    views = video.find_element(by='xpath', value='.//*[@id="metadata-line"]/span[1]').text
    link = video.find_element(by='xpath', value='.//*[@id="video-title-link"]').get_attribute("href")
    
    videos_list.append({
        "title": title,
        "views": views,
        "link": link,
    })
#     print(f"{idx+1}. ",title, views, link, "\n")
```

```python
driver.quit()
```

```python
df = pd.DataFrame(videos_list)
df
```
