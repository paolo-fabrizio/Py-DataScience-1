```python
import pandas as pd
```

```python
# leyendo el archivo csv
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams.head()
```

# 1 Operaciones

## 1.1 Operaciones en columnas

```python
# suma total de una sola columna
df_exams['math score'].sum()
```

```python
# contar, promedio, desv. estamdar, maximo y mínimo
print(df_exams['math score'].sum())
print(df_exams['math score'].mean())
print(df_exams['math score'].std())
print(df_exams['math score'].max())
print(df_exams['math score'].min())
```

```python
# calcula todas las operaciones anteriores más rapido con .describe()
df_exams.describe()
```

## 1.2 Operaciones en filas

```python
# calcular la suma en una fila
df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']
```

```python
# calcular el score promedio y asignar los resultados a una nueva columna
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score'])/3
```

# 2 Método Value Counts

```python
# Contar los elementos en la columna "gender"
#funcion len
print(len(df_exams['gender']))

#metodo count()
print(df_exams['gender'].count())
```

```python
# Contar los elementos en la columna "gender" por categoría
df_exams['gender'].value_counts()
```

```python
# obtener la frecuencia relativa
df_exams['gender'].value_counts(normalize=True)
```

```python
# contar elementos de la columna "parental level of education" por categoría
df_exams['parental level of education'].value_counts()
```

```python
# obtener la frecuencia relativa y redondear a 2 decimales
df_exams['parental level of education'].value_counts(normalize=True).round(2)
```

# 3 Ordenar un Dataframe con sort_values()

```python
# ordenar por una columna
df_exams.sort_values('math score')
```

```python
# orden descendiente por una columna
df_exams.sort_values('math score', ascending=False)
```

```python
# orden descendiente por múltiples columnas
# la prioridad para ordenar es siempre del primer elemento
df_exams.sort_values(['math score', 'reading score'], ascending=False)
```

```python
# usando el parámetro "implace" para actualizar el dataframe
# los métodos de ordenamiento no actualizan el dataframe por defecto

df_exams.sort_values(['math score', 'reading score'], ascending=False, inplace=True)
df_exams
```

```python
# orden descendiente con una función key
# usamos la funcion lambda para poner en minúscula los valores de gender y
# poder compararlas entre sí
df_exams.sort_values('race/ethnicity', ascending=True,
                    key=lambda col:col.str.lower())
df_exams
```
