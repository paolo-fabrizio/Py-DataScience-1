# 3 Agregar una nueva columna

## 1.1 Agregar una nueva columna con un valor

```python
# agregar una nueva columna al dataframe
df_exams['language score'] = 70
df_exams
```

```python
# importando numpy
import numpy as np
```

```python
# crear un array de 1000 elementos
language_array = np.arange(0, 1000)
len(language_array)
```

```python
# agregar una nueva columna al dataframe con un array
df_exams['language score'] = language_array
df_exams
```

```python
# agregando aleatoriedad a language_array con numeros entre el 1 y 100
int_language_array = np.random.randint(1, 100, size=1000)
len(int_language_array)
```

```python
# el mínimo del método randint es incluyente, mientras que el máximo es excluyente
min(int_language_array)
max(int_language_array)
```

```python
# actualizando columna 'language score' con el array de aleatorios
df_exams['language score'] = int_language_array
df_exams
```

```python
# crear numeros decimales aleatorios entre 1 y 100
np.random.uniform(1, 100, size=1000)
```
